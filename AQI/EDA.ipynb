## Task 1: Finding which city has the lowest NaN values

```python
data = df.set_index(df["City"]).isna().sum(level=0)
data
```

Only calculating the NaN values is not appropriate. We first need to calculate the total data points and then calculate the percentage of not-null points.

```python
df["City"].value_counts()
```

Dropping unnecessary columns and creating columns for all the data points per city

```python
total_values = [48192,2680,22784,29269,48192,6903,22468,7263,48192,9229,48192,3852,40258,12002,48107,26705,28025,3854,19503,
                48192,48192,44554,7402,22161,26651,35053]
data["total_val"] = total_values
```

```python
data.drop(["City", "Datetime", "NO", "NOx", "Benzene", "Toluene", "Xylene"], axis=1, inplace=True)
```

```python
data
```

We will calulate the percentage of not null values for each city by dividing No of rows which has values by Total No of rows

```python
data["true_PM2.5"] = ((data["total_val"] - data["PM2.5"]) / data["total_val"]) * 100
data["true_PM10"] = ((data["total_val"] - data["PM10"]) / data["total_val"]) * 100
data["true_NO2"] = ((data["total_val"] - data["NO2"]) / data["total_val"]) * 100
data["true_NH3"] = ((data["total_val"] - data["NH3"]) / data["total_val"]) * 100
data["true_CO"] = ((data["total_val"] - data["CO"]) / data["total_val"]) * 100
data["true_SO2"] = ((data["total_val"] - data["SO2"]) / data["total_val"]) * 100
data["true_O3"] = ((data["total_val"] - data["O3"]) / data["total_val"]) * 100
```

```python
data
```
